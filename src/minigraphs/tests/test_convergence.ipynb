# Test greedy convergence

```python
# Preamble
n_vertices = 100
p = 0.1
beta = np.inf
epsilon = 0.05

targets = {
    "assortativity": 0.5
}

# Instantiate annealer
annealer = mini.MH(metrics_funcs)

# Instantiate ER graph
G = nx.erdos_renyi_graph(n_vertices,p)

# Optimize graph
annealer.transform(G,targets,beta=np.inf,epsilon=0.001)
trajectories = annealer.trajectories_

axes = mini.MH.plot_trajectories(trajectories,targets)
```
