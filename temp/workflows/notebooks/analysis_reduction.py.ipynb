```python
# start coding here
import matplotlib.pyplot as plt 
import matplotlib 
import pandas as pd
import yaml

def load_dict(file):
    with open(file,'r') as f:
        return yaml.safe_load(f)
    

# Update matplotlib rc params
for param, value in snakemake.params.rc.items():
    matplotlib.rcParams[param] = value
    
# Load dataframes
df_names = ["parameters",
            "graph_metrics",
            "distribution_degree",
            "distribution_distance",
            "qois_sir"]

dfs = {}
for name, file in zip(df_names, snakemake.input):
    if "qois" in name:
        header = [0,1]
    else:
        header = 0
        
    # Load Dataframe
    dfs[name] =  pd.read_csv(file,index_col=0,header=header)
    
    if name == "parameters":
        targets = dfs[name].iloc[:,0]
        names_parameters = dfs[name].columns
    elif "qois" in name:
        dfs[name] = dfs[name].groupby(targets).mean()
    else:
        dfs[name] = dfs[name].groupby(targets).describe(percentiles=[])

names = ["graph_metrics","distribution_degree","distribution_distance"]
files = [snakemake.input.metrics,
         snakemake.input.d_degree,
         snakemake.input.d_distance]
parent = {name: load_dict(file) for name, file in zip(names,files)}
parent['graph_metrics']

method = snakemake.wildcards.method
        
```

```python
fig, axes = plt.subplots(2,1,figsize=(6,3))
# DATA
df = dfs["graph_metrics"]
p = parent["graph_metrics"]
x = df.index 

# PLOT
axes[0].errorbar(x,(p["n_nodes"]-df["n_nodes"]["mean"])/p["n_nodes"])
axes[1].errorbar(x,(p["n_edges"]-df["n_edges"]["mean"])/p["n_edges"])

# FORMAT
axes[0].set_title("Fraction of nodes removed")
axes[0].set_xticks([])


axes[1].set_title("Fraction of edges removed")
axes[1].set_xlabel(names_parameters[0].capitalize())

for i, ax in enumerate(axes):
    if method == "sparsification":
        ax.set_xscale("log")
    
    if i != 1:
        ax.set_xticks([])
        
    ax.set_ylim([-0.1,1.1])
            

plt.tight_layout()


metrics = ["density",
           "clustering",
           "assortativity",
           "connectivity"]

fig, axes = plt.subplots(len(metrics),1,figsize=(7.5,15))
for i, metric in enumerate(metrics):
    axes[i].errorbar(x, df[metric]["mean"])
    axes[i].set_ylabel(metric.capitalize())
    
    if method == "sparsification":
        axes[i].set_xscale("log")
    
plt.tight_layout()
plt.savefig(snakemake.output[0])
```
