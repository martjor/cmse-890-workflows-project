```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd

matplotlib.rc_file('.matplotlibrc')
metric='density'
```

```python
metrics_file = f"../results/{metric}/metrics.csv"

metrics_df = pd.read_csv(metrics_file,index_col=0)
metrics_group = metrics_df.groupby('target').describe()
metrics_group
```

```python
stats_file = f"../results/{metric}/stats_degree.csv" 
stats_degree_df = pd.read_csv(stats_file,index_col=0)
stats_degree_group = stats_degree_df.groupby(metrics_df['target']).describe()
stats_degree_group
```

```python
fig, axes = plt.subplots(1,2,figsize=(6,3))
axes[0].errorbar(metrics_group.index,
                 metrics_group['connectivity']['mean'])
axes[1].errorbar(metrics_group.index,metrics_group['n_components']['mean'])
if metric == 'density': 
    ax.set_xscale('log')
```

```python
fig, axes = plt.subplots(1,2,figsize=(6,3))
mean = stats_degree_group['mean']['mean']
axes[0].errorbar(stats_degree_group.index,mean)
axes[1].errorbar(stats_degree_group.index,
                 mean/stats_degree_group['m_2']['mean'])
```
