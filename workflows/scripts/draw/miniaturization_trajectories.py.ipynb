```python
# Import modules
from minigraphs.inspection import plot_loss, plot_losses, plot_schedules
from scripts.utils.io import load_dict
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Read weights
weights = {metric: weight for metric, weight in load_dict(snakemake.input[1]).items() if metric != "beta"}

# Read target metrics
targets = {metric: target for metric, target in load_dict(snakemake.input[0]).items() if metric in weights.keys()}

# Read trajectories
dfs = []
for index in range(6):
    dfs.append(
        pd.read_parquet(
            f"{snakemake.input[2]}/replica_{index}.parquet"
        )
    )
    
fig, ax = plt.subplots(3,1,figsize=(8,12), dpi=300)

# Find best replica 
losses = [df['loss'].iat[-1] for df in dfs]
idx_best = losses.index(min(losses))

plot_losses(dfs, ax[0])
plot_loss(dfs[idx_best], targets, weights, ax[1])
plot_schedules(dfs, ax[2])

plt.tight_layout()
plt.savefig(snakemake.output[0], bbox_inches='tight')
    
    
```
