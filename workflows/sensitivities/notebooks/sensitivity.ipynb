```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

metric = "density"
plt.rcParams["figure.autolayout"] = True
plt.rcParams["figure.dpi"] = 300
plt.rcParams["lines.linewidth"] = 0.5
plt.rcParams["lines.marker"] = 'o'
plt.rcParams["lines.markersize"] = 1
plt.rcParams["errorbar.capsize"] = 2
plt.rcParams["axes.labelsize"]=10
plt.rcParams["font.size"]=7
```

# Load Data

```python
DIR = f"../results/{metric}/"

graphs_df = pd.read_csv(DIR + "graphs.csv", index_col=0)
graphs_group = graphs_df.groupby('target').describe()
graphs_group
```

```python
dynamics_df = pd.read_csv(DIR + "qois_sir.csv",index_col=0,header=[0,1])
dynamics_group = dynamics_df.groupby(graphs_df['target']).mean()
dynamics_group
```

# Analysis

```python
fig, ax = plt.subplots(figsize=(3,3))

y = graphs_group[metric]['mean']
yerr = graphs_group[metric]['std'] * 3
ax.errorbar(graphs_group.index,y,yerr=yerr,capthick=0.5)
ax.set_xlabel('p')
ax.set_ylabel(metric.capitalize())
ax.set_title(f"{metric.capitalize()} as a function of generation parameter")

# Adjust scale for density
if metric == "density":
    ax.set_xscale('log')
    ax.set_yscale('log')
```

```python
fig, axes = plt.subplots(1,2,figsize=(6,3))
fig.suptitle("Metrics as a function of target")

x = graphs_group.index 
other_metrics = list(graphs_group.columns.get_level_values(0).unique())
other_metrics.remove(metric)

for i, ax in enumerate(axes):
    y = graphs_group.loc[:,other_metrics[i]]['mean']
    yerr = graphs_group.loc[:,other_metrics[i]]['std']*3
    
    ax.errorbar(x,y,yerr=yerr,capthick=0.5)
    ax.set_ylabel(other_metrics[i].capitalize())
    ax.set_xlabel(metric.capitalize())
    
    if metric == "density":
        ax.set_xscale('log')
    
    if other_metrics[i] == "assortativity":
        ax.set_ylim([-1,1])
    else:
        ax.set_ylim([0,1])
```

```python
fig, axes = plt.subplots(1,3,figsize=(9,3))
fig.suptitle("SIR dynamics as a function of target metric")

x = graphs_group.index
metrics = dynamics_group.columns.get_level_values(0).unique()
for i, ax in enumerate(axes):
    df = dynamics_group.loc[:,metrics[i]]
    y = dynamics_group.loc[:,metrics[i]]['mean']
    lomin = dynamics_group.loc[:,metrics[i]]
    ax.errorbar(x,y)
    
    ax.set_xlabel(metric.capitalize())
    ax.set_ylabel(metrics[i].capitalize())
    
    if metric == "density":
        ax.set_xscale("log")
```
