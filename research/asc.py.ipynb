```python
import matplotlib.pyplot as plt
import numpy as np 

def metropolis(x, beta=1.0):
    return np.minimum(np.exp(-beta * x), 1.0)

fig, ax = plt.subplots(figsize=(7,5), dpi=300)

x = np.linspace(-2,2,200)
red = plt.get_cmap('autumn')(0.0)
blue = plt.get_cmap('winter')(0.0)

ax.axvline(0.0,color='k',linestyle=':')
hot = metropolis(x,1.0)
ax.plot(x, hot, color=red, label=r'$\beta=1.0$')
ax.fill_between(x, hot, color=red, alpha=0.3)

ax.plot(x, metropolis(x,10.0), color=blue, label=r'$\beta=10.0$')
ax.fill_between(x, metropolis(x,10.0), color=blue, alpha=0.3)

ax.set_ylim([0,2.0])
ax.set_xlim([-1.0,1.0])
ax.set_yticks([0.0,0.5,1.0])
ax.set_xticks([-1.0,-0.5,0.0,0.5,1.0])

ax.legend()

ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.savefig('metropolis.png',transparent=True)
```

```python
fig, ax = plt.subplots(figsize=(20,5), dpi=300)

x = np.linspace(0,10,100)
y = 1 / (1 + np.exp(1*(x-5)))

ax.plot(x,y,color='magenta',label=r'1$/\beta$')
ax.set_yticks([0,1])
ax.set_yticklabels(['0',r'$\infty$'])
ax.set_xticks([0,10])
ax.set_xticklabels([r'$t_0$',r'$t_{\max}$'])
ax.legend()

ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.savefig('schedule.png', transparent=300)
```
